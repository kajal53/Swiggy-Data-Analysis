```python
df = pd.read_csv('swiggy_3000_orders_25_columns.csv')
```

```python
df['Coupon_Code'] = df['Coupon_Code'].fillna('None')
df['Feedback'] = df['Feedback'].fillna('No Feedback')
# Convert the 'Delivery_Time' column to string type before using .str accessor
df['Delivery_Time'] = df['Delivery_Time'].astype(str).str.replace(' min', '').astype(int)
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
city_orders = df['City'].value_counts().reset_index()
city_orders.columns = ['City', 'Order_Count']
```

```python
item_orders = df['Item'].value_counts().reset_index()
item_orders.columns = ['Item', 'Order_Count']
```

```python
avg_delivery_time = df.groupby(['City', 'Weather'])['Delivery_Time'].mean().reset_index()
```

```python
# Restaurant ratings vs. order volume
restaurant_stats = df.groupby('Restaurant_Name').agg({
    'Order_ID': 'count',
    'Rating': 'mean'
}).reset_index()
restaurant_stats.columns = ['Restaurant_Name', 'Order_Count', 'Avg_Rating']
```

```python
# Cancellation rate by payment method
cancel_rate = df.groupby('Payment_Method')['Delivery_Status'].apply(
    lambda x: (x == 'Cancelled').mean() * 100
).reset_index()
cancel_rate.columns = ['Payment_Method', 'Cancellation_Rate']
```

```python
# Step 4: Visualizations
# Plot 1: Order Distribution by City
# Use a different color scale for unique colors per city
fig1 = px.bar(city_orders, x='City', y='Order_Count', title='Order Distribution by City',
              color='City', text='Order_Count', color_continuous_scale=px.colors.qualitative.Set3) # Using Set3 qualitative scale
fig1.update_traces(textposition='outside')
fig1.update_layout(xaxis_title='City', yaxis_title='Number of Orders', showlegend=False)
fig1.show()  # Display in Colab
fig1.write_html('city_orders.html')  # Save for download

# Import the files module from google.colab to enable file downloads
# from google.colab import files
# files.download('city_orders.html')  # Download the HTML file
```

```python
# Plot 2: Order Distribution by Item
fig2 = px.pie(item_orders, values='Order_Count', names='Item', title='Order Distribution by Item',
              color_discrete_sequence=px.colors.qualitative.Pastel)
fig2.update_traces(textinfo='percent+label')
fig2.show()  # Display in Colab
fig2.write_html('item_orders.html')
files.download('item_orders.html')
```

```python
# Plot 3: Average Delivery Time by City and Weather
fig3 = px.bar(avg_delivery_time, x='City', y='Delivery_Time', color='Weather',
              title='Average Delivery Time by City and Weather', barmode='group')
fig3.update_layout(xaxis_title='City', yaxis_title='Average Delivery Time (min)')
fig3.show()  # Display in Colab
fig3.write_html('delivery_time.html')
files.download('delivery_time.html')
```

```python
# Plot 4: Restaurant Ratings vs. Order Volume
fig4 = px.scatter(restaurant_stats, x='Order_Count', y='Avg_Rating', size='Order_Count',
                  color='Restaurant_Name', text='Restaurant_Name',
                  title='Restaurant Ratings vs. Order Volume')
fig4.update_traces(textposition='top center')
fig4.update_layout(xaxis_title='Number of Orders', yaxis_title='Average Rating')
fig4.show()  # Display in Colab
# fig4.write_html('restaurant_stats.html')
# files.download('restaurant_stats.html')
```

```python
fig5 = go.Figure(data=[
    go.Bar(x=cancel_rate['Payment_Method'], y=cancel_rate['Cancellation_Rate'],
           text=cancel_rate['Cancellation_Rate'].round(2), textposition='auto')
])
fig5.update_layout(title='Cancellation Rate by Payment Method',
                   xaxis_title='Payment Method', yaxis_title='Cancellation Rate (%)')
fig5.show()  # Display in Colab
# fig5.write_html('cancel_rate.html')
# files.download('cancel_rate.html')
```

```python
# NEW Plot 6: Orders Per Month (Line Chart)
df['Order_Month'] = df['Order_Date'].dt.to_period('M') # Extract the month (and year)
orders_per_month = df['Order_Month'].value_counts().sort_index().reset_index() # Count orders per month and sort
orders_per_month.columns = ['Month', 'Order_Count']
orders_per_month['Month'] = orders_per_month['Month'].astype(str) # Convert Month back to string for plotting

# NEW Plot 6: Orders Per Month (Line Chart)
fig6 = px.line(orders_per_month, x='Month', y='Order_Count', title='Orders Per Month',
               markers=True, line_shape='linear')
fig6.update_layout(xaxis_title='Month', yaxis_title='Number of Orders')
fig6.show()
# fig6.write_html('orders_per_month.html')
# files.download('orders_per_month.html')
```

```python
# Before Plot 7, add the following code to calculate delivery status distribution:
delivery_status = df['Delivery_Status'].value_counts().reset_index()
delivery_status.columns = ['Delivery_Status', 'Order_Count']
```

```python
# NEW Plot 7: Delivery Status Distribution (Pie Chart)
# Before Plot 7, add the following code to calculate delivery status distribution:
delivery_status = df['Delivery_Status'].value_counts().reset_index()
delivery_status.columns = ['Delivery_Status', 'Order_Count']

fig7 = px.pie(delivery_status, values='Order_Count', names='Delivery_Status',
              title='Delivery Status Distribution',
              color_discrete_sequence=px.colors.qualitative.Set2)
fig7.update_traces(textinfo='percent+label')
fig7.show()
# fig7.write_html('delivery_status.html')
# files.download('delivery_status.html')
```
